```python
from IPython.display import display
import pandas as pd

from src import config
```

```python
data_dir = config.dir_data / "icecat-taxonomy"
```

```python
taxonomy_cases = pd.read_excel(data_dir + "mobile_phone_cases.xlsx")
taxonomy_cases
```

```python
taxonomy_phones = pd.read_excel(data_dir + "smartphones.xlsx")
taxonomy_phones
```

```python
ids_cases = set(taxonomy_cases["Feature_id"])
ids_phones = set(taxonomy_phones["Feature_id"])

ids = ids_cases.union(ids_phones)
ids_both = ids_cases.intersection(ids_phones)

print("unique ids:", len(ids))
print("shared ids:", len(ids_both))
```

```python
features_cases = taxonomy_cases[["Feature_id", "Feature", "Feature group", "Type", "Is mandatory"]]

column_name = {
    "Feature_id": "id",
    "Feature": "name_cases",
    "Feature group": "group_cases",
    "Type": "type_cases",
    "Is mandatory": "required_cases",
}

features_cases = features_cases.rename(columns=column_name)
features_cases
```

```python
features_phones = taxonomy_phones[["Feature_id", "Feature", "Feature group", "Type", "Is mandatory"]]

column_name = {
    "Feature_id": "id",
    "Feature": "name_phones",
    "Feature group": "group_phones",
    "Type": "type_phones",
    "Is mandatory": "required_phones",
}

features_phones = features_phones.rename(columns=column_name)
features_phones
```

```python
features = pd.merge(features_cases, features_phones, 'outer')
features
```

```python
for field_name in ["name", "group", "type"]:
    key_cases = f"{field_name}_cases"
    key_phones = f"{field_name}_phones"

    tmp = features[
        features[key_cases].notna() & features[key_phones].notna() &
        (features[key_cases] != features[key_phones])
    ]
    print(f"--- {field_name} -------------------")
    display(tmp[["id", key_cases, key_phones]])
```

```python
tmp = features[
    features["group_cases"].notna() & features["group_phones"].notna() &
    (features["group_cases"] != features["group_phones"])
]

tmp[["id", "name_cases", "group_cases", "group_phones"]]
```

```python
columns = ["Feature_id", "Feature", "Feature group", "Type", "Is mandatory"]
column_names = {
    "Feature_id": "id",
    "Feature": "name",
    "Feature group": "group",
    "Type": "type",
    "Is mandatory": "required_cases",
}

features_cases = taxonomy_cases[columns]
features_cases = features_cases.rename(columns=column_names)

column_names["Is mandatory"] = "required_phones"

features_phones = taxonomy_phones[columns]
features_phones = features_phones.rename(columns=column_names)

features = pd.merge(features_cases, features_phones, "outer")
features
```

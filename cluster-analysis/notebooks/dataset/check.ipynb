```python
attr_dict = cache.attributes_dict


def cast_prods(prods: list[akeneo.Product]) -> dict[tuple[str, str], list]:
    result = {
        ("__", "identifier"): [],
        ("__", "family"): [],
        ("__", "categories"): [],
    }

    for i in range(len(prods)):
        prod = prods[i]
        result[("__", "identifier")].append(prod.identifier)
        result[("__", "family")].append(prod.family)
        result[("__", "categories")].append(prod.categories)

        for attr_code in prod.values.keys():
            attr = attr_dict[attr_code]
            key = (attr.group, attr.name)
            if key not in result:
                result[key] = [None] * i

            value = (
                attr.get_option(prod.values[attr_code]).labels
                if attr.type
                in {
                    akeneo.AttributeType.SELECT_SINGLE,
                    akeneo.AttributeType.SELECT_MULTI,
                }
                else prod.values[attr_code]
            )

            result[key].append(value)

        for key, value in result.items():
            if len(value) <= i:
                result[key].append(None)

    sorted_result = {}
    for key, value in sorted(result.items()):
        sorted_result[key] = value
    return sorted_result


df = pd.DataFrame(cast_prods(cache.products))
df
```

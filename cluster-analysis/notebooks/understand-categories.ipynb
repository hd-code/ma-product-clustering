```python
import pandas as pd

import akeneo as ak
```

```python
connector, _ = ak.create_from_env()
```

```python
categories = connector.get_categories()
products = connector.get_products()
```

```python
cats = pd.DataFrame(categories)
cats["num_of_products"] = 0
cats
```

```python
for p in products:
    for c in p.categories:
        cats.loc[cats["code"] == c, "num_of_products"] += 1
cats
```

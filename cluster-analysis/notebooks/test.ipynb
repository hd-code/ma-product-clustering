```python
def translate(obj: dict): return "__"
config = Config(type_hooks={akeneo.LocalStr: translate}, cast=[int, float, str])
measurements = [from_dict(akeneo.MeasurementFamily, m, config) for m in measurements]
df = pd.DataFrame(measurements)
df
```

```python

import os

from IPython.display import display
import pandas as pd

from akeneo import AkeneoClient, AkeneoConnector, create_client_from_env
from config import data_dir
```

```python
csv_dir = data_dir / "0-exploration"
try:
    os.mkdir(csv_dir)
except:
    pass
```

```python
client: AkeneoClient = create_client_from_env()
connector = AkeneoConnector(client, "en_US")
```

```python
lists = [
    ("attributes", connector.get_attributes),
    ("categories", connector.get_categories),
    ("families", connector.get_families),
    ("product-models", connector.get_product_models),
    ("products", connector.get_products),
]

for name, getter in lists:
    values = getter()
    df = pd.DataFrame(values)
    df.to_csv(csv_dir / f"{name}.csv", index=False)
    display(f"--- {name} ---------------------------")
    display(df.head())
```

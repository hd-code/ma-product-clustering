```python
lists_no_args = routes[routes["is_list"] & routes["path_vars"].isin([[]])]
lists_no_args
```

```python
csv_dir = data_dir / "0-akeneo-api"
csv_dir.mkdir(parents=True, exist_ok=True)

def print_and_safe_csv(name: str, data):
    df = pd.DataFrame(data)
    df.to_csv(csv_dir / f"{name}.csv", index=False)

    display(f"--- {name} -----------------------")
    display(df.head())
```

```python
for _, row in lists_no_args[["id", "path"]].iterrows():
    route_id = row[0]
    name = str(row[1]).split("/")[-1]

    res = client.get_list(route_id)
    print_and_safe_csv(name, res)

    if name == "product-models":
        data = [{"code": entry["code"], **entry["values"]} for entry in res]
        print_and_safe_csv("product-models_values", data)
    elif name == "products":
        data = [{"identifier": entry["identifier"], **entry["values"]} for entry in res]
        print_and_safe_csv("products_values", data)
```

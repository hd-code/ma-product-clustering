```python
csv_dir = data_dir / "02-akeneo-data-cleaned"
shutil.rmtree(csv_dir, ignore_errors=True)
csv_dir.mkdir(parents=True, exist_ok=True)

cases = [
    ("attributes", connector.get_attributes),
    ("categories", connector.get_categories),
    ("families", connector.get_families),
    ("products", connector.get_products),
]

for name, func in cases:
    res = func()
    df = pd.DataFrame(res)
    print(f"--- {name} -----------------------------")
    display(df)
    df.to_csv(csv_dir / f"{name}.csv", index=False)

    if name == "products":
        res = [prod.values for prod in res]
        df = pd.DataFrame(res)
        print(f"--- {name}__values -----------------------------")
        display(df)
        df.to_csv(csv_dir / f"{name}__values.csv", index=False)
```
